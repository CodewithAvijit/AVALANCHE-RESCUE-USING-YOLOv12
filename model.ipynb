VIDEO PROCESS

```python
tracker = DeepSort(max_age=30, n_init=2, nms_max_overlap=1.0, max_cosine_distance=0.3)

cap = cv2.VideoCapture(r"D:\AVALANCHE_RESCUE_USING_IMAGE_PROCESSING\TEST\testvideo\Avalanche Survival 101.mp4")

while True:
    ret, frame = cap.read()
    if not ret:
        break
    results = model(frame, conf=0.2, imgsz=640)
    detections = []
    for boxes, cls, conf in zip(results[0].boxes.xyxy.cpu().numpy(),
                                results[0].boxes.cls.cpu().numpy(),
                                results[0].boxes.conf.cpu().numpy()):
        x1, y1, x2, y2 = map(int, boxes)
        w, h = x2 - x1, y2 - y1
        class_name = model.names[int(cls)]
        detections.append(([x1, y1, w, h], conf, class_name))

    tracks = tracker.update_tracks(detections, frame=frame)

    for track in tracks:
        if not track.is_confirmed():
            continue
        track_id = track.track_id
        ltrb = track.to_ltrb()  # left, top, right, bottom
        x1, y1, x2, y2 = map(int, ltrb)
        label = f"ID {track_id}"
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, f"HUMAN: {label}", (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)

    cv2.imshow("RESCUECAM", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
model = YOLO(r"D:\AVALANCHE_RESCUE_USING_IMAGE_PROCESSING\MODEL\rescue_model\weights\best.pt")
cap = cv2.imread(r"D:\AVALANCHE_RESCUE_USING_IMAGE_PROCESSING\search-and-rescue\valid\images\gss202_jpg.rf.2982de26ecefcc175d2d5bbb44393c4d.jpg")
results = model(cap, conf=0.2, imgsz=640)

colors = {}
def get_color(cls):
    if cls not in colors:
        colors[cls] = [random.randint(50, 255) for _ in range(3)]
    return colors[cls]

for boxes, cls, conf in zip(results[0].boxes.xyxy, results[0].boxes.cls, results[0].boxes.conf):
    x1, y1, x2, y2 = map(int, boxes)
    cls_name = model.names[int(cls)]
    confidence = conf * 100
    color = get_color(cls_name)

    cv2.rectangle(cap, (x1, y1), (x2, y2), color, 4, lineType=cv2.LINE_AA)

    label = f"{cls_name} {confidence:.1f}%"
    (tw, th), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_DUPLEX, 0.8, 2)

    cv2.rectangle(cap, (x1, y1 - th - 10), (x1 + tw + 10, y1), color, -1)

    cv2.putText(cap, label, (x1 + 5, y1 - 5), cv2.FONT_HERSHEY_DUPLEX,
                0.8, (0, 0, 0), 3, lineType=cv2.LINE_AA)  
    cv2.putText(cap, label, (x1 + 5, y1 - 5), cv2.FONT_HERSHEY_DUPLEX,
                0.8, (255, 255, 255), 1, lineType=cv2.LINE_AA)

cv2.imshow("RESCUE_CAM", cap)
cv2.imwrite(r"D:\AVALANCHE_RESCUE_USING_IMAGE_PROCESSING\testimage\test2.jpg", cap)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
